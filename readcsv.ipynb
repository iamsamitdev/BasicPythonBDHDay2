##### Step 4: วาดกราฟด้วย plotly

```python
import plotly.express as px

fig = px.line(df, x='AAPL_x', y='AAPL_y', title='Apple Share Prices 2014')
fig.show()
```

```python
fig = px.bar(df, x='AAPL_x', y='AAPL_y', title='Apple Share Prices 2014')
fig.show()
```

```python
fig = px.area(df, x='AAPL_x', y='AAPL_y', title='Apple Share Prices 2014')
fig.show()
```

```python
fig = px.scatter(df, x='AAPL_x', y='AAPL_y', title='Apple Share Prices 2014')
fig.show()
```

##### Step 5: การอ่านข้อมูลตัวอย่างจาก plotly และทำการ plot graph

```python
import plotly.express as px

wind_df = px.data.wind()

fig = px.line_polar(
    wind_df, 
    r='frequency', 
    theta='direction', 
    color='strength', 
    line_close=True,
)
fig.show()
```

```python
iris_df = px.data.iris()
# iris_df
fig = px.scatter(iris_df, x='sepal_width', y='sepal_length', color='species')
fig.show()
```
